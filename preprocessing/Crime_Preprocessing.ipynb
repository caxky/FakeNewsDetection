```python
from datasets import load_dataset
import pandas as pd
```

```python
def pd_set_see_full_dataframe():
    # Permanently changes the pandas settings
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    # pd.set_option('display.max_colwidth', -1)

def pd_reset_options():
    # Resets the options
    pd.reset_option('all')

def display_whole_df(df):
    pd_set_see_full_dataframe()
    display(df)
    pd_reset_options()
```

## Imports and Merge


## Fa-kes


```python
df_fa_kes = pd.read_csv('../data/crime/fa-kes/FA-KES-Dataset.csv',encoding='latin-1')
df_fa_kes.head()
```

```python
df_fa_kes = df_fa_kes[['article_title','article_content','labels','date','location', 'source']]
```

```python
df_fa_kes.head()
```

```python
display(df_fa_kes['labels'].value_counts())
df_fa_kes['labels'].replace({0:False, 1:True}, inplace=True)
df_fa_kes['labels'].replace({False:1, True:0}, inplace=True)
display(df_fa_kes.head())
display(df_fa_kes['labels'].value_counts())
```

```python
df_fa_kes.rename(columns={'article_title':'title','article_content':'text', 'labels':'label'}, inplace=True)
display(df_fa_kes.head())
```

```python
df_fa_kes = df_fa_kes[['text', 'label']].assign(metadata=df_fa_kes.iloc[:,2:].agg(dict,1))
display(df_fa_kes.head())
```

```python
df_fa_kes.to_feather('../data/FA-KES-Dataset.feather')
```

## Snopes



```python
df_emergent = pd.read_csv('../data/crime/snopes/snopes_phase2_clean_2018_7_3.csv')
df_emergent.head()
```

```python
pd_set_see_full_dataframe()
display(df_emergent['article_category_phase1'].value_counts())
pd_reset_options()
```

```python
df_test = df_emergent[df_emergent['article_category_phase1'].str.contains("Crime")==True]
```

```python
df_test['article_category_phase1'].value_counts()
```

```python
df_test.head()
```

```python
df_test = df_test.sort_values(by=['article_category_phase1'])
df_test = df_test.drop_duplicates(keep='last')
```

```python
df_test['fact_rating_phase1'].value_counts()
```

```python
# df_test['fact_rating_phase1'].replace('mostly false', 'false', inplace=True)
df_test['fact_rating_phase1'].replace('mostly true', 'true', inplace=True)
df_test['fact_rating_phase1'].replace('unproven', 'true', inplace=True)
```

```python
df_snopes = df_test[df_test['fact_rating_phase1'].isin(['false', 'true'])]
```

```python
df_snopes['fact_rating_phase1'].value_counts()
```

```python
df_snopes = df_snopes.rename(columns={'fact_rating_phase1':'label','article_title_phase1':'text'})
```

```python
df_snopes = df_snopes[['text', 'label']].assign(metadata=df_snopes.iloc[:,2:].agg(dict,1))
```

```python
df_snopes['label'].replace({'false':1, 'true':0}, inplace=True)
```

```python
df_snopes.head()
```

```python
df_snopes['label'].value_counts()
```

```python
df_snopes.to_feather('../data/snope.feather')
```
