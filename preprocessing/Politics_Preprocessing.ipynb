```python
from datasets import load_dataset
import pandas as pd
```

```python
def pd_set_see_full_dataframe():
    # Permanently changes the pandas settings
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    # pd.set_option('display.max_colwidth', -1)

def pd_reset_options():
    # Resets the options
    pd.reset_option('all')

def display_whole_df(df):
    pd_set_see_full_dataframe()
    display(df)
    pd_reset_options()
```

#### Liar Dataset (HuggingFace)

Labels
* '0': 'false'
* '1': half-true
* '2': mostly-true
* '3': 'true'
* '4': barely-true
* '5': pants-fire

New 
* Fake (false, barely-true, pants)= 1
* True (half-true, mostly-true, true)= 0


```python
df_liar_dataset['label'].value_counts()
```

```python
df_liar_dataset['label'] = df_liar_dataset['label'].replace([0,4,5], -1)
df_liar_dataset['label'] = df_liar_dataset['label'].replace([1,2,3], 0)
df_liar_dataset['label'] = df_liar_dataset['label'].replace([-1], 1)
```

```python
df_liar_dataset = df_liar_dataset.rename(columns={'statement': 'text', 'speaker': 'author'})
df_liar_dataset = df_liar_dataset[[ 'text', 'label', 'author',
 'subject',
 'job_title',
 'state_info',
 'party_affiliation',
 'context']]
# Create metadata
df_liar_dataset = df_liar_dataset[['text', 'label',]].assign(metadata=df_liar_dataset.iloc[:,2:].agg(dict,1))
```

```python
df_liar_dataset.head()
```

#### Fake News Dataset (Already F = 1, T = 0)
labels
* 1: unreliable
* 0: reliable

```python
df_fake_news_dataset = df_fake_news_dataset[['text', 'label', 'author','title']]
```

```python
# meta data
df_fake_news_dataset = df_fake_news_dataset[['text', 'label']].assign(metadata=df_fake_news_dataset.iloc[:,2:].agg(dict,1))
```

```python
df_fake_news_dataset.head()
```

#### ISOT Dataset (Labels done when importing )

labels 
* 0 == true 
* 1 == false

```python
df_isot_dataset['author'] = None
df_isot_dataset = df_isot_dataset[['text', 'label', 'author', 'title', 'subject', 'date']]
# Create metadata column 
df_isot_dataset = df_isot_dataset[['text', 'label',]].assign(metadata=df_isot_dataset.iloc[:,2:].agg(dict,1))
df_isot_dataset.head()
```

#### Politifact Dataset
* label: The verdict of fact check in one of 6 categories {true, mostly-true, half-true, mostly-false, false, and pants-fire}

```python
df_politifact_dataset = df_politifact_dataset.rename(columns={'verdict': 'label', 'statement_originator':'author', 'statement': 'text'})
```

```python
df_politifact_dataset['label'].value_counts()
```

```python
df_politifact_dataset['label'] = df_politifact_dataset['label'].replace(['true', 'half-true', 'mostly-true'], 0)
df_politifact_dataset['label'] = df_politifact_dataset['label'].replace(['false', 'pants-fire', 'mostly-false'], 1)
```

```python
df_politifact_dataset['label'].value_counts()
```

```python
df_politifact_dataset = df_politifact_dataset[['text', 'label', 'author',
 'statement_date',
 'statement_source',
 'factchecker',
 'factcheck_date',
 'factcheck_analysis_link']]
df_politifact_dataset = df_politifact_dataset[['text', 'label']].assign(metadata=df_politifact_dataset.iloc[:,2:].agg(dict,1))
df_politifact_dataset.head()
```
